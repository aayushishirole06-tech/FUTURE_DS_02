```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df.info()
```

```python
churn_counts = df['Churn'].value_counts()

plt.figure(figsize=(5,5))
plt.pie(churn_counts,
        labels=churn_counts.index,
        autopct='%1.1f%%')

plt.title('Customer Churn Distribution')
plt.show()
```

```python
contract_churn = pd.crosstab(df['Contract'], df['Churn'])

contract_churn.plot(kind='bar', figsize=(7,5))

plt.title('Churn by Contract Type')
plt.xlabel('Contract Type')
plt.ylabel('Number of Customers')
plt.xticks(rotation=0)
plt.show()
```

```python
payment_churn = pd.crosstab(df['PaymentMethod'], df['Churn'])

payment_churn.plot(kind='bar', figsize=(9,5))

plt.title('Churn by Payment Method')
plt.xlabel('Payment Method')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df[df['Churn']=='Yes']['tenure'],
         bins=30)

plt.title('Tenure of Churned Customers')
plt.xlabel('Months Stayed')
plt.ylabel('Customers')
plt.show()
```

```python
internet_churn = pd.crosstab(df['InternetService'], df['Churn'])

internet_churn.plot(kind='bar', figsize=(7,5))

plt.title('Churn by Internet Service')
plt.xlabel('Internet Service')
plt.ylabel('Number of Customers')
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df[df['Churn']=='No']['MonthlyCharges'],
         alpha=0.6,
         label='Stayed')

plt.hist(df[df['Churn']=='Yes']['MonthlyCharges'],
         alpha=0.6,
         label='Churned')

plt.title('Monthly Charges vs Churn')
plt.xlabel('Monthly Charges')
plt.ylabel('Customers')
plt.legend()
plt.show()
```

```python
senior_churn = pd.crosstab(df['SeniorCitizen'], df['Churn'])

senior_churn.plot(kind='bar', figsize=(6,5))

plt.title('Churn by Senior Citizen')
plt.xlabel('Senior Citizen (0 = No, 1 = Yes)')
plt.ylabel('Customers')
plt.xticks(rotation=0)
plt.show()
```
